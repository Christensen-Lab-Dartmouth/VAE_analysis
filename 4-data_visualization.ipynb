```python
subtypes = ['Normal', 'LumA', 'LumB', 'Her2', 'Basal']
methyl_merged.loc[methyl_merged['PAM50.RNAseq'].isin(subtypes)]

def f(row):
    if row['PAM50.RNAseq'] == 'Normal':
        val = 0
    elif row['PAM50.RNAseq'] == 'Basal':
        val = 1
    elif row['PAM50.RNAseq'] == 'Her2':
        val = 2
    elif row['PAM50.RNAseq'] == 'LumA':
        val = 3
    elif row['PAM50.RNAseq'] == 'LumB':
        val = 4
    else:
        val = -1
    return val

methyl_merged['PAM50.numerical'] = methyl_merged.apply(f, axis=1)
methyl_merged.shape
```

```python
from scipy import stats
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D 
from matplotlib import colors
from matplotlib import rc
import matplotlib

x = methyl_merged[['1_x']]
y = methyl_merged[['2_x']]
z = methyl_merged[['3_x']]
cmap = colors.ListedColormap(['k','b','y','r','g'])


fig = plt.figure(figsize=(10, 10))

ax = fig.add_subplot(111, projection='3d')
ax.scatter(x, z, y, c=methyl_merged['PAM50.numerical'], marker='o', cmap=cmap, edgecolors='none')

# Set title
#plt.title('3D t-SNE dimensionality reduction')

ax.set_xlabel('t-SNE Dim 1')
ax.set_ylabel('t-SNE Dim 3')
ax.set_zlabel('t-SNE Dim 2')

scatter1_proxy = matplotlib.lines.Line2D([0],[0], linestyle="none", marker = 'o', c='k')
scatter2_proxy = matplotlib.lines.Line2D([0],[0], linestyle="none", marker = 'o', c='b')
scatter3_proxy = matplotlib.lines.Line2D([0],[0], linestyle="none", marker = 'o', c='y')
scatter4_proxy = matplotlib.lines.Line2D([0],[0], linestyle="none", marker = 'o', c='r')
scatter5_proxy = matplotlib.lines.Line2D([0],[0], linestyle="none", marker = 'o', c='g')
#scatter6_proxy = matplotlib.lines.Line2D([0],[0], linestyle="none", marker = 'o', c='w')
ax.legend([scatter1_proxy, scatter2_proxy, scatter3_proxy, scatter4_proxy, scatter5_proxy], 
          ['Normal', 'Basal-like', 'Her2', 'LumA', 'LumB'], 
          numpoints = 1, bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

ax.w_xaxis.set_pane_color((1.0, 1.0, 1.0, 1.0))
ax.w_yaxis.set_pane_color((1.0, 1.0, 1.0, 1.0))
ax.w_zaxis.set_pane_color((1.0, 1.0, 1.0, 1.0))

rc('font',size=15)
rc('font',family='serif')

plt.savefig('figures/tSNE_3D_figure_vae_300K_100latent_.eps', format='eps', dpi=300, bbox_inches='tight')

plt.show()
```
